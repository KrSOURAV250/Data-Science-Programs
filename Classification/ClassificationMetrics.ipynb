```python
df = pd.read_csv(f"The Data\BreastCancer.csv")
df
```

```python
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xTrain = scaler.fit_transform(xTrain)
xTest = scaler.transform(xTest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xTrain, yTrain)
```

```python
yPred = knn.predict(xTest)
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(yTest, yPred))
```

### Precision

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(yTest, yPred, pos_label='B')
```

```python
pd.DataFrame(precision_score(yTest, yPred, average=None))
```

```python
precision_score(yTest, yPred, average="macro")
```

```python
precision_score(yTest, yPred, average="weighted")
```

### Recall

```python
recall_score(yTest, yPred, pos_label='M')
```

```python
pd.DataFrame(recall_score(yTest, yPred, average=None))
```

```python
recall_score(yTest, yPred, average="macro")
```

```python
recall_score(yTest, yPred, average="weighted")
```

### F1 Score

```python
f1_score(yTest, yPred,pos_label='B')
```

```python
pd.DataFrame(f1_score(yTest, yPred, average=None))
```

```python
f1_score(yTest, yPred, average="macro")
```

```python
f1_score(yTest, yPred, average="weighted")
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(yTest, yPred))
```
