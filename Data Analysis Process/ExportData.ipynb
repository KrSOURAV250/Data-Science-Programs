### Pandas Export
- to csv
- to excel
- to html
- to json
- to sql

```python
import pandas as pd
```

### to_csv

```python
df = pd.read_csv(r"The Data\deliveries.csv")
df.head()
```

```python
temp_df = df.groupby("batsman")[["batsman_runs"]].sum()
temp_df
```

```python
temp_df.to_csv("batsman_runs.csv")
```

```python
temp_df = df.groupby("batsman")["batsman_runs"].sum().reset_index()
temp_df
```

```python
temp_df.to_csv("batsman_runs.csv",index=False)
```

```python
df.pivot_table(index="batsman", columns="bowling_team", values="batsman_runs", aggfunc="sum").to_csv("batsman_vs_team.csv")
```

### to_excel

```python
df
```

```python
temp_df = df.groupby("batsman")["batsman_runs"].sum().reset_index()
temp_df
```

```python
temp_df.to_excel("batsman_runss.xlsx")
```

```python
temp_df
```

```python
temp_df2 = df.pivot_table(
    index="batsman", columns="bowling_team", values="batsman_runs", aggfunc="sum")
temp_df2
```

```python
with pd.ExcelWriter("output.xlsx") as writer:
    temp_df.to_excel(writer,sheet_name="sheet 101")
    temp_df2.to_excel(writer,sheet_name="sheet 102")
```

### to_html

```python
temp_df = df.query("batsman_runs == 6").pivot_table(
    index="over", columns="ball", values="batsman_runs", aggfunc="count")
temp_df
```

```python
temp_df.to_html("sises_heatmap.html")
```

### to_json

```python
temp_df = df.groupby(["batting_team", "batsman"])[
    "batsman_runs"].sum().unstack()
temp_df
```

```python
temp_df.to_json("ipl.json")
```
