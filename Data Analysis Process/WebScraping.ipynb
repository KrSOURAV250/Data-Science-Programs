```python
name = []
rating = []
reviews = []
pubpvt=[]
location=[]
yearold=[]
emp=[]
about=[]
for i in company:
    name.append(i.find("h2").text.strip())
    rating.append(i.find("p", class_="rating").text.strip())
    reviews.append(i.find('a', class_="review-count").text.strip())
    pubpvt.append(i.find_all('p', class_="infoEntity")[0].text.strip())
    location.append(i.find_all('p', class_="infoEntity")[1].text.strip())
    yearold.append(i.find_all('p', class_="infoEntity")[2].text.strip())
    emp.append(i.find_all('p', class_="infoEntity")[3].text.strip())
    about.append(soup.find('p', class_="description").text.strip())
```

```python
name = []
rating = []
reviews = []
pubpvt = []
location = []
yearold = []
emp = []
about = []
for i in range(1, 11):
    header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36"}
    webPage = requests.get(
        rf"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={i}", headers=header).text
    soup = BeautifulSoup(webPage, "lxml")
    company = soup.find_all("div", class_="company-content-wrapper")
    for j in company:
        name.append(j.find("h2").text.strip())
        rating.append(j.find("p", class_="rating").text.strip())
        reviews.append(j.find('a', class_="review-count").text.strip())
        pubpvt.append(j.find_all('p', class_="infoEntity")[0].text.strip())
        try:
            location.append(j.find_all(
                'p', class_="infoEntity")[1].text.strip())
        except:
            location.append(np.nan)
        try:
            yearold.append(j.find_all('p', class_="infoEntity")
                           [2].text.strip())
        except:
            yearold.append(np.nan)
        try:
            emp.append(j.find_all('p', class_="infoEntity")[3].text.strip())
        except:
            emp.append(np.nan)
        try:
            about.append(j.find('p', class_="description").text.strip())
        except:
            about.append(np.nan)
```

```python
d = {"name": name, "rating": rating, "reviews": reviews,
     "type": pubpvt, "hq": location, "Old": yearold,"about":about}
pd.DataFrame(d)
```
