```python
import pandas as pd
```

```python
df = pd.read_csv(r"The Data\Human Activity Recognition with Smartphones train.csv").drop(columns="subject")
df
```

```python
df["Activity"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = df.drop("Activity", axis=1)
y = df["Activity"]

# Encode Target Label
le = LabelEncoder()
y= le.fit_transform(y)

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
xTrain.shape, xTest.shape
```

```python
# Applying Logistic Regression 
logReg = LogisticRegression(max_iter=1000)
logReg.fit(xTrain,yTrain)
yPred = logReg.predict(xTest)
```

```python
accuracy_score(yTest, yPred)
```

```python
def get_duplicate_columns(df):
    
    duplicate_columns = {}
    seen_columns = {}

    for column in df.columns:
        current_column = df[column]

        # Convert column data to bytes
        try:
            current_column_hash = current_column.values.tobytes()
        except AttributeError:
            current_column_hash = current_column.to_string().encode()

        if current_column_hash in seen_columns:
            if seen_columns[current_column_hash] in duplicate_columns:
                duplicate_columns[seen_columns[current_column_hash]].append(column)
            else:
                duplicate_columns[seen_columns[current_column_hash]] = [column]
        else:
            seen_columns[current_column_hash] = column

    return duplicate_columns
```

```python
duplicateColumns = get_duplicate_columns(xTrain)
duplicateColumns
```

```python
for i in duplicateColumns.values():
    xTrain.drop(columns=i, inplace=True)
    xTest.drop(columns=i, inplace=True)
```

### 3. Correlation

```python
import plotly.express as px

px.imshow(xTrain.corr(), aspect="auto")
```

```python
corrMatrix = xTrain.corr()

columns = corrMatrix.columns
columnsToDrop = []
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        if corrMatrix.loc[columns[i], columns[j]] > 0.95:
            columnsToDrop.append(columns[j])
columnsToDrop = set(columnsToDrop)
len(columnsToDrop)
```

```python
xTrain.drop(columns=columnsToDrop, inplace=True)
xTest.drop(columns=columnsToDrop, inplace=True)
```

### 5. Chi-Square Test

```python
titanic = pd.read_csv(r"The Data\Titanic train.csv")[["Pclass", "Sex", "SibSp", "Parch", "Embarked", "Survived"]]
titanic
```

```python
ct = pd.crosstab(titanic["Survived"], titanic["Sex"])
ct
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(ct)
```

```python
score = []
for feature in titanic.columns[:-1]:
    ct = pd.crosstab(titanic["Survived"], titanic[feature])

    pValue = chi2_contingency(ct)[1]
    score.append(pValue)
```

```python
score
```

```python
px.bar(pd.Series(score, index=titanic.columns[:-1]).sort_values(ascending=True))
```

```python
# Using Sklearn 
from sklearn.feature_selection import chi2

# assuming titanic is your DataFrame and 'Survived' is the target column

# Encode categorical variables
le = LabelEncoder()
titanic_encoded = titanic.apply(le.fit_transform)

X = titanic_encoded.drop('Survived', axis=1)
y = titanic_encoded['Survived']

# Calculate chi-squared stats
chi_scores = chi2(X, y)

# chi_scores[1] are the p-values of each feature.
p_values = pd.Series(chi_scores[1], index = X.columns)
p_values.sort_values(inplace = True)

px.bar(p_values)
```
