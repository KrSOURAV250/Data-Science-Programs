```python
plt.figure(figsize=(15,6))
plt.scatter(batters["avg"], batters["strike_rate"])
plt.title("Something")
plt.xlabel("avg")
plt.ylabel("strike_rate")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax.scatter(batters["avg"], batters["strike_rate"])
ax.set_title("Something")
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 6), sharex=True)
ax[0].scatter(batters["avg"], batters["strike_rate"], color="red")
ax[1].scatter(batters["avg"], batters["runs"])
ax[0].set_title("Avg vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 10))
ax[0, 0].scatter(batters["avg"], batters["strike_rate"], color="red")
ax[1, 0].hist(batters["avg"])
ax[1, 1].hist(batters["runs"])
ax[0, 0].set_title("Avg vs Strike Rate")
ax[0, 0].set_ylabel("Strike Rate")
ax[0, 1].scatter(batters["avg"], batters["runs"])
ax[0, 1].set_title("Avg vs Runs")
ax[0, 1].set_xlabel("Average")
ax[0, 1].set_ylabel("Runs")
```

### 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
xx.shape, yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure()
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

### Contour Plots

```python
z = xx**2 + yy**2
fig = plt.figure()
ax = plt.subplot()
p = ax.contour(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

```python
fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

### HeatMap

```python
delivery = pd.read_csv(r"The Data\IPL_Ball_by_Ball_2008_2022.csv")
delivery.head()
```

```python
delivery["ballnumber"].unique()
```

```python
temp_df = delivery[(delivery["ballnumber"].isin([1, 2, 3, 4, 5, 6]))
                   & (delivery["batsman_run"] == 6)]
temp_df.head()
```

```python
grid = temp_df.pivot_table(
    index="overs", columns="ballnumber", values="batsman_run", aggfunc="count")
grid
```

```python
plt.figure(figsize=(10, 7))
plt.imshow(grid)
plt.xticks(np.arange(0, 6), np.arange(1, 7))
plt.yticks(delivery["overs"].unique(), np.arange(1, 21))
plt.colorbar()
```
