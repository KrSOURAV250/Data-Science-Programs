```python

plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="dashed")
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashed")

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="dotted")
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dotted")

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="dashdot")
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashdot")

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="solid", linewidth=3)
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashdot", linewidth=2)

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# markers
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="solid", linewidth=3, marker="D")
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashdot", linewidth=2, marker="D")

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="solid", linewidth=3, marker="+",markersize=10)
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashdot", linewidth=2, marker="o", markersize=10)

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# legend location
plt.plot(batsman["index"], batsman["V Kohli"],
         color="green", linestyle="solid", linewidth=3, marker="+",markersize=10, label="Virat")
plt.plot(batsman["index"], batsman["RG Sharma"],
         color="black", linestyle="dashdot", linewidth=2, marker="o", markersize=5, label="Rohit")

plt.title("Rohit Sharma vs Virat Kohli Career Comparison")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.legend(loc="best")
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simplr bar chart
childern = [10, 20, 40, 10, 30]
colors = "red,blue,green,yellow,pink".split(',')
plt.bar(colors,childern, color="black")
```

```python

# horizontal bar chart
plt.barh(colors, childern)
```

```python
# color and label
df = pd.read_csv(r"The Data\batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df["2015"], width=0.2)
plt.bar(np.arange(df.shape[0]), df["2016"], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df["2017"], width=0.2)
# xticks
plt.xticks(np.arange(df.shape[0]), df["batsman"])
```

```python
# xticks -> rotation
plt.bar(np.arange(df.shape[0]) - 0.2, df["2015"], width=0.2)
plt.bar(np.arange(df.shape[0]), df["2016"], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df["2017"], width=0.2)
plt.xticks(np.arange(df.shape[0]), df["batsman"])
plt.xticks(rotation=45)
plt.show()
```

```python
df
```

```python
# stacked bar chart
plt.bar(df["batsman"], df["2017"], label="2017")
plt.bar(df["batsman"], df["2016"], bottom=df["2017"], label="2016")
plt.bar(df["batsman"], df["2015"], bottom=df["2016"] + df["2017"], label=2015)
plt.legend()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23, 45, 100, 20, 49]
subjects = "eng,science,maths,sst,hindi".split(',')
plt.pie(data, labels=subjects)
plt.show()
```

```python

# dataset
df = pd.read_csv(r"The Data\gayle-175.csv")
df
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%0.1f%%")
plt.show()
```

```python
# explode shadow
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%0.1f%%", explode=[0.2,0,0,0,0,0], shadow=True)
plt.show()
```
