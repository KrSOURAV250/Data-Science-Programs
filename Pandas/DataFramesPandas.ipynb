### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all final winers
ipl[ipl["MatchNumber"] == "Final"][["Season", "WinningTeam"]]
```

```python
# how many super over finishes have occurs
(ipl["SuperOver"] == 'Y').sum()
```

```python
# how many matches has CSK won in kolkata
((ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")).sum()
```

```python
(ipl["TossWinner"] == ipl["WinningTeam"]).sum() * 100 / ipl.shape[0]
```

```python
# movies with rating higher than 8 and votes > 10000
movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"] > 10000)
       ].loc[:, ["title_x", "imdb_rating", "imdb_votes"]].head()
```

```python
# Action Movies with rating higher than 7.5
movies[(movies["genres"].str.split("|").apply(
    lambda l:"Action" in l)) & (movies["imdb_rating"] > 7.5)][["title_x", "genres", "imdb_rating"]].head()
```

```python
movies[(movies["genres"].str.contains("Action")) & (
    movies["imdb_rating"] > 7.5)][["title_x", "genres", "imdb_rating"]].head()
```

```python
# Write a function that can return record of 2 team against each other
def record(teamA, teamB):
    return ipl[((ipl["Team1"] == teamA) & (ipl["Team2"] == teamB)) | ((ipl["Team1"]
                                                                       == teamB) & (ipl["Team2"] == teamA))][["Team1", "Team2", "WinningTeam"]]
```

```python
record("Punjab Kings", "Mumbai Indians")
```
