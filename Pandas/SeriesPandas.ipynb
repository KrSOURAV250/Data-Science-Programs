### What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

https://pandas.pydata.org/about/index.html

### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### Importing Pandas

```python
import pandas as pd
import numpy as np
import sys
```

### Series From List

```python
# String
country = ["India", "Pakistan", "USA", "Nepal", "Sri Lanka"]
pd.Series(country)
```

```python
# Integers
runs = [13, 24, 56, 78, 100]
runs = pd.Series(runs)
runs
```

```python
# custom index
marks = [67, 57, 89, 100]
sub = ["Maths", "English", "Science", "Punjabi"]
pd.Series(marks, index=sub)
```

```python
# setting a name
marks = pd.Series(marks, index=sub, name="My Marks")
marks
```

### Series From Dictionary

```python
marks = {"Maths": 67, "English": 57, "Science": 89, "Punjabi": 100}
marks_series = pd.Series(marks, name="My Marks")
```

### Boolean Indexing on Series

```python
# Find no. of 50's and 100's scored by virat kohli
vk[(vk == 50) | (vk == 100)].size
```

```python
# find number of ducks
vk[vk == 0].size
```

```python
# count no. of days when i had more than 2000 subs a day 
subs[subs > 200].size
```

```python
# find actors who have more than 20 movies
movies.value_counts()[movies.value_counts() > 20]
```

### Some Important Series Methods

- astype

- isnull

```python
# astype
vk, sys.getsizeof(vk)
```

```python
vk.astype(np.int8), sys.getsizeof(vk.astype(np.int8))
```

```python
# between
vk.between(51,99)
```

```python
vk[vk.between(51,99)].size
```

```python
# clip
subs.clip(100, 200)
```

```python
# drop_duplicates
temp = pd.Series([1, 1, 2, 2, 3, 3, 4, 4, 5, 6, 7, 8, 9])
temp
```

```python
temp.drop_duplicates(), temp.drop_duplicates(keep="last")
```

```python
temp.duplicated().sum()
```

```python
vk.duplicated().sum()
```

```python
# isnull
temp = pd.Series([1, 2, 3, np.nan, 5, 6, np.nan, 8, np.nan, 10])
temp
```

```python
temp.size, temp.count() # size count all the values, count() skip nan values.
```

```python
temp.isnull().sum()
```

```python
# dropna
temp.dropna()
```

```python
# fillna
temp.fillna(250), temp.fillna(temp.mean())
```

```python
# isin 
vk.isin([49,99]), vk[vk.isin([49,99,199])]
```

```python
# apply
movies.apply(lambda x : x.split()[0].upper())
```

```python
subs, subs.mean()
```

```python
subs.apply(lambda x: "Good Day" if x > subs.mean() else "Bad Day")
```
