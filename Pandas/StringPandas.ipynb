```python
# problem in vectorizzed operations in vanilla python
s = ["cat", "mat", None, "rat"]
[i.startswith('c') for i in s]
```

```python
# How pandas solve this issue?
s = pd.Series(["cat", "mat", None, "rat"])
# string accessor
s.str.startswith('c')
# it is fast and optimized
```

```python
df = pd.read_csv(r"TheData\titanic.csv")
df.head()
```

```python
df["Name"]
```

```python
# Common Functions
# lower/upper/capitalize/title
df["Name"].str.title()
```

```python
# len 
df["Name"].str.len()
```

```python
df["Name"].str.len().max()
```

```python
df["Name"][df["Name"].str.len()==82].values[0]
```

```python
# strip
df["Name"].str.strip()
```

```python
# split -> get
df["LastName"] = df["Name"].str.split(',').str.get(0)
```

```python
df.head()
```

```python
df[["title", "firstname"]] = df["Name"].str.split(',').str.get(
    1).str.strip().str.split(' ', n=1, expand=True)
df.head()
```

```python
df["title"].value_counts()
```

```python
df["title"] = df["title"].str.replace("Ms.", "Miss.")
df["title"] = df["title"].str.replace("Mlle.", "Miss.")
```

```python
df["title"].value_counts()
```

```python
# filtering
# startswith/endswith
df[df["firstname"].str.startswith('A')]
```

```python
# isdigit/isalpha
df[df["firstname"].str.isdigit()]
```

```python
# Applying regular expression
# contains
# search jhon -> both case
df[df['firstname'].str.contains('john',case=False)].head()
```

```python
# find lastnames with start and end char vovel
df[df["LastName"].str.contains("^[aeiouAEIOU].+[aeiouAEIOU]$")]["LastName"]
```

```python
# find lastnames with start and end char conconents
df[df["LastName"].str.contains("^[^aeiouAEIOU].+[^aeiouAEIOU]$")]["LastName"]
```

```python
# slicing
df['Name'].str[:4]
df['Name'].str[::2]
df['Name'].str[::-1]
```
