```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"The Data\placement.csv")
df.head()
```

```python
fig = px.scatter(df, x="cgpa", y="package", title="Nothing TOO", labels={
                 "cgpa": "CGPA", "package": "Package(in LPA)"}, trendline="ols")
fig.update_traces(marker=dict(size=5), selector=dict(mode='markers'))
```
