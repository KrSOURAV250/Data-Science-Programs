```python
import pandas as pd
import numpy as np
```

```python
class MyLinearRegression:
    def __init__(self) -> None:
        self.b = None
        self.m = None

    def fit(self, xTrain, yTrain):
        num = 0
        den = 0
        xTrainAvg = xTrain.mean()
        yTrainAvg = yTrain.mean()
        for i in range(xTrain.shape[0]):
            num += ((xTrain[i] - xTrainAvg) * (yTrain[i] - yTrainAvg))
            den += (xTrain[i] - xTrainAvg) ** 2
        self.m = num / den
        self.b = yTrain.mean() - (self.m * xTrain.mean())
        return "MyLinearRegression"

    def predict(self, xTest):
        return (self.m * xTest) + self.b
```
